```python
import pandas as pd
#Read in the master plus blast database!
master_df = pd.read_csv('/stor/work/Marcotte/project/drbarth/asgard/membrane_proteome/data/processed/Asgard_Guaymas_MasterDatabase_plus_blast.csv', index_col=0)
master_df.head()
```

```python
#Function that takes in a 3line file and outputs a pandas dataframe containing the protein names, the corresponding AA sequence, and prediction from TMBed
def read3line(filepath):    
    protein_names = []
    sequence = []
    prediction = []
    with open(filepath, "r") as a_file:
        for line in a_file:
            if line.startswith('>'):
                #split_line = line.strip('\n')
                split_line = line.split('>')
                split_line = split_line[1].split('\n')
                prot_name = split_line[0]
                protein_names.append(prot_name)
                for a in range(2):
                    if a == 0:
                        seq = next(a_file)
                        split = seq.split('\n')
                        sequence.append(split[0])
                    else: 
                        pred = next(a_file)
                        pred_split = pred.split('\n')
                        prediction.append(pred_split[0])
    df = pd.DataFrame(list(zip(protein_names, sequence, prediction)),
               columns =['ProteinID', 'aa_sequence', 'TMBed_prediction'])
    return df

```

```python
asgard_tmbed_fp = '/stor/work/Marcotte/project/drbarth/asgard/membrane_proteome/data/raw/Guaymas_Asgard_deeptmhmmpredictedprots2_test.fasta.tmbed.3lines'
asgard_tmbed_df = read3line(asgard_tmbed_fp)
#Drop the aa_sequence column for simplicity
tmbed_df = asgard_tmbed_df.drop(columns=['aa_sequence'])

asgard_tmbed_df.head()
```

```python
#Merge TMBed predictgions with the master dataframe base on the ProteinID column 
master_plus_tmbed = master_df.merge(tmbed_df, on='ProteinID', how='left')
master_plus_tmbed.head()
```
